```
#@title Install requirements & Load library
%%capture
!pip install -U easynmt
!pip install polib
!pip install tqdm

import io
import polib
from easynmt import EasyNMT
from tqdm.notebook import tqdm

model = EasyNMT('opus-mt')
po = polib.pofile(path)
```

```
#@title Starting traduction :) Enjoy it! 😁😉
po.metadata['Language-Team'] = 'JonatanGk'
po.metadata['X-Generator'] = 'REPO LINK HERE'
for lang in languages.split(','):
  po.metadata['Language'] = lang
  for entry in tqdm(po, desc = 'Translation to %s' % lang):
    entry.msgstr = model.translate(entry.msgid, source_lang='en', target_lang=lang)
  path_out = '/content/%s.po' % lang
  po.save(path_out)
  files.download(path_out)
```
